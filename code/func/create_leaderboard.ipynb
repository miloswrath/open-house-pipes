```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df =pd.read_csv('./../data/None.csv')
test = df[df['condition'] == 'test']
test
```

```python
# get total responses that are not "None" by each block (1, 2)

block1 = test[test['block_c'] == 1]
block2 = test[test['block_c'] == 2]
count1 = 0
for i in range(len(block1)):
    if block1['response'].iloc[i] != 'undefined':
        count1 += 1
    
count2 = 0
for i in range(len(block2)):
    if block2['response'].iloc[i] != 'undefined':
        count2 += 1

total_responses = [[count1, len(block1)], [count2, len(block2)]]
total_responses



```

```python
# get total correct responses by each block (1, 2)
correct = [[block1['correct'].astype('int').sum(), len(block1)], [block2['correct'].astype('int').sum(), len(block2)]]
correct
```

```python
# get average response time by each block (1, 2)
avg_time = [[block1['response_time'].mean(), len(block1)], [block2['response_time'].mean(), len(block2)]]

#get min and max response time by each block (1, 2)
min_time = [[block1['response_time'].min(), len(block1)], [block2['response_time'].min(), len(block2)]]
max_time = [[block1['response_time'].max(), len(block1)], [block2['response_time'].max(), len(block2)]]

rt = pd.DataFrame(avg_time, columns = ['avg_time', 'total'])
rt['block'] = [1, 2]
rt['min_time'] = [min_time[0][0], min_time[1][0]]
rt['max_time'] = [max_time[0][0], max_time[1][0]]

rt['accuracy'] = [correct[0][0]/correct[0][1], correct[1][0]/correct[1][1]]
rt
```

```python
composite = abs(1/rt['accuracy'] - rt['avg_time'] * rt['total'] - (rt['max_time'] - rt['min_time']))
composite = (composite[0] + composite[1])/2
composite = composite.round(0)
composite = composite.astype('int')

composite
```

```python
scores = pd.DataFrame(columns = ['composite', 'sub_name'])

scores['composite'] = [composite]
scores['sub_name'] = df['multichar_response'].iloc[0]
scores
```
